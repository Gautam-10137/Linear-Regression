```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("Advertising.csv")
```

```python
df.head()
```

```python
df['total_spend']=df['TV']+df['radio']+df['newspaper']
```

```python
df.head()
```

```python
# regression plot in seaborn
sns.regplot(data=df,x='total_spend',y='sales')
```

# making regression plot manually by finding exact equation

```python
X=df['total_spend']
y=df['sales']
```

```python
#  built in method to find coefficients using least square method in numpy   -> np.polyfit()
# y=mx+b
# y=B1x+B0
```

```python
np.polyfit(X,y,deg=1)  # -> return an array of coefficient values [B1,B0]
```

```python
potential_spend=np.linspace(0,500,100)
```

```python
predicted_sales=0.04868788*potential_spend+4.24302822
```

```python
sns.scatterplot(x='total_spend',y='sales',data=df)
plt.plot(potential_spend,predicted_sales,color='red')
```

# finding relationship b/w higher order equations

```python
np.polyfit(X,y,3)
# y= B3(x**3)+B2(x**2)+B1(x)+B0
```

```python
pot_spend=np.linspace(0,500,100)
```

```python
pred_sales=3.07615033e-07*(pot_spend**3)+-1.89392449e-04*(pot_spend**2)+ 8.20886302e-02*(pot_spend)+2.70495053e+00
```

```python
sns.scatterplot(x='total_spend',y='sales',data=df)
plt.plot(pot_spend,pred_sales,color='red')
```
