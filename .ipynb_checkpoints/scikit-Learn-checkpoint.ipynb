```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("Advertising.csv")
```

```python
df.head()
```

```python
sns.pairplot(df)
```

```python
X=df.drop('sales',axis=1)
```

```python
X.head()
```

```python
y=df['sales']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=101)
```

```python
len(df)
```

```python
len(X_test)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
```

```python
model.fit(X_train,y_train)
```

```python
test_predictions=model.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
```

```python
mean_absolute_error(y_test,test_predictions)
#  thsi error doesnot take into account far off data points
```

```python
df['sales'].mean()
```

```python
#  so, we have 10% of error
```

```python
# root mean square error  RMSE    -> more efficient
np.sqrt(mean_squared_error(y_test,test_predictions))
```

```python
#  finding residual error
test_residuals=y_test- test_predictions
```

```python
sns.scatterplot(x=y_test,y=test_residuals)
plt.axhline(y=0,color='r',ls='--')
#  if we see a pattern in this , then linear regression is not applicable
```

```python
sns.displot(x=test_residuals,bins=20,kde=True)
```

```python
#  probability plot
import scipy as sp
```

```python
fig,ax=plt.subplots(figsize=(6,8),dpi=200)
_=sp.stats.probplot(test_residuals,plot=ax)
```

## Training Final model ( since we have assumed that this algorithm is giving good result)

```python
final_model=LinearRegression()
```

```python
final_model.fit(X,y) # fitting whole data
```

```python
# showing coefficients
final_model.coef_   # coefficients are just relationships:
#  note this is how much fraction of increase in sales with increase in 1 unit of these features
# e.g. 1 unit increase in TV increase 0.045 of sales (by assuming radio & newspaper spend as fixed)
```

```python
df.head()
```

```python
#  plotting graph between predictive sales & actual sales
y_hat=final_model.predict(X)
```

```python
fig,axes=plt.subplots(figsize=(12,3))
axes.plot(df['TV'],df['sales'],'o')
axes.plot(df['TV'],y_hat,'o')
axes.set_ylabel('sales')
axes.set_title('TV spend')
#  visualizing these for all feautures become less useful, so we use residual plot
```

# Finally deploying the model

```python
from joblib import dump,load
```

```python
dump(final_model,'final_sales_model.joblib')
```

```python
loaded_model=load('final_sales_model.joblib')
```
